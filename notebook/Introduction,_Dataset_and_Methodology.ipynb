## Background:
Each year thousands of people are murdered across the United States and many cases remain unsolved. Violent crimes are a danger to everyone in society as they threaten the trust and security of every individual. Though violent crimes on a per capita basis are in decline, the percentage of unsolved murder cases is increasing. This can be attributed to several factors as criminals become savvier in how they commit crimes and information about obscuring evidence can be easily found online. Besides the clearance rate of identifying offenders decreasing, evidence shows that policing agencies also carry bias in terms of how the cases are handled. This investigation aims to identify patterns (like victimology, geography, and crime characteristics to name a few) and use them to review these crimes with the ultimate goals of holding law enforcement agencies accountable and resolving unsolved cases.


# Datasets:
The Murder Accountability Project (MAP) (http://www.murderdata.org/) is composed of active and retired law enforcement personnel who are trying to reduce the number of unsolved murders in the United States. The organization was founded to bring attention to these and to ensure that law enforcement agencies are taking the care and attention needed to achieve justice. Using freedom of information requests and publicly available data, MAP maintains a database of homicides across the entire country dating back to 1965 and are consistently updated with new information as police departments release it. Over eight hundred thousand records detail about the victims, murderers (when known), policing agencies, locations, and crime details, are catalogued in the project. The most comprehensive dataset contains 31 different predictor variables including the victim’s race, age, sex, location as well as the murder’s information if the case had been solved (Figure 1. Full details: https://www.dropbox.com/s/lo6tgo8nnbpqeru/MAPdefinitionsSHR.pdf?dl=1).

The offender's characteristics are modelled after the four algorithms individually. The Gaussian Naive Bayes model performs the worst because it has the highest error for each characteristic. It was especially bad for predicting the age category as there is significant variance within these category. The result was improved slightly by binning the age into 5 year regions but this resulted in very limited improvement in testing error. The random forest model shows an improvement on each category predicted so the parameters were looked at further and optimized for enhancement. The neural network is comparable to the optimized random forest but there is overfitting in the forest. The best model to determine an offender's profile is the neural network because the training and testing errors are minimal with a low mean squared error for age. Also, there is no indication of under or over fitting the model since the training and testing errors are similar. We would also be cautious about using methods such as Bayes as the model relies on prior probabilities. Recent news shows there are growing numbers of cases that suggest bias' exist within police deparments as they may be targetting minorities unfairly. Building a profiling model such as this based on those skewed results may continue this bias unfairly.  Using a neural net would still be subject to these inputs but it would have more sophisticatation and resampling to limit these effects.
